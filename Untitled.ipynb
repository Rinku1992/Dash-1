```python
######### Functions #########
def rescaled_points(xlist,ylist,sites):
    sites_scaled = sites/111000
    centrexlist = [x[0] for x in sites_scaled.tolist()]
    centreylist = [x[1] for x in sites_scaled.tolist()]
    centrextranslated = [(x+min(xlist)) for x in centrexlist]
    centreytranslated = [(x+min(ylist)) for x in centreylist]
    centrepointzip = zip(centrextranslated,centreytranslated)
    centrepoints = np.array(list(centrepointzip))
    return(centrepoints)
```

```python
def generate_candidate_sites(points,M=100):
    '''
    Generate M candidate sites with the convex hull of a point set
    Input:
        points: a Numpy array with shape of (N,2)
        M: the number of candidate sites to generate
    Return:
        sites: a Numpy array with shape of (M,2)
    '''
    hull = ConvexHull(points)
    polygon_points = points[hull.vertices]
    poly = Polygon(polygon_points)
    min_x, min_y, max_x, max_y = poly.bounds
    sites = []
    while len(sites) < M:
        random_point = Point([random.uniform(min_x, max_x),
                             random.uniform(min_y, max_y)])
        if (random_point.within(poly)):
            sites.append(random_point)
    return np.array([(p.x,p.y) for p in sites])
#############################################
```

```python
def mclp(points,K,radius,M):
    """
    Solve maximum covering location problem
    Input:
        points: input points, Numpy array in shape of [N,2]
        K: the number of sites to select
        radius: the radius of circle
        M: the number of candidate sites, which will randomly generated inside
        the ConvexHull wrapped by the polygon
    Return:
        opt_sites: locations K optimal sites, Numpy array in shape of [K,2]
        f: the optimal value of the objective function
    """
    import time
    start = time.time()
    sites = generate_candidate_sites(points,M)
    J = sites.shape[0]
    I = points.shape[0]
    D = distance_matrix(points,sites)
    mask1 = D<=radius
    D[mask1]=1
    D[~mask1]=0
    # Build model
    m = Model()
    # Add variables
    x = {}
    y = {}
    for i in range(I):
      y[i] = m.addVar(vtype=GRB.BINARY, name="y%d" % i)
    for j in range(J):
      x[j] = m.addVar(vtype=GRB.BINARY, name="x%d" % j)

    m.update()
    # Add constraints
    m.addConstr(quicksum(x[j] for j in range(J)) == K)

    for i in range(I):
        m.addConstr(quicksum(x[j] for j in np.where(D[i]==1)[0]) >= y[i])

    m.setObjective(quicksum(y[i]for i in range(I)),GRB.MAXIMIZE)
    m.setParam('OutputFlag', 0)
    m.optimize()
    end = time.time()

    solution = []
    if m.status == GRB.Status.OPTIMAL:
        for v in m.getVars():
            # print v.varName,v.x
            if v.x==1 and v.varName[0]=="x":
               solution.append(int(v.varName[1:]))
    opt_sites = sites[solution]
    return opt_sites,m.objVal
########################################################################################
```

```python
def extract_coverage(opt_sites,allpoints,radius):
    points = allpoints
    sitedict = {}
    pointdict = {}
    for cntr,sitepoint in enumerate(list(opt_sites)):
        sitedict['p'+str(cntr)] = list(sitepoint)
        pointdict['p'+str(cntr)] = []
    
    
    allcoveredlist = []
    for point in list(points):
        p1 = (point[0],point[1])
        for k,sitepoint in sitedict.items():
            p2 = (sitepoint[0],sitepoint[1])
            radialdist = distance.euclidean(p1, p2)
            if radialdist <= radius:
                pointdict[k].append(list(p1))
                allcoveredlist.append(list(p1))
    
    not_coveredlist = []
    for p in points:
        ptemp = list((p[0],p[1]))
        if ptemp in allcoveredlist:
            continue
        not_coveredlist.append(list(ptemp))
    return allcoveredlist, not_coveredlist,sitedict
################################################################################################
```

```python
def cluster_optimization(clusterlist,micro_radius,femto_radius):
    random.seed(2019)
    df = pd.DataFrame()
    stat_list_micro = dict()
    stat_list_femto = dict()
    for ctr,area in enumerate(clusterlist):
        print("\n\n*************Running Optimization for cluster region "+str(ctr))
        location1= area[['Latitude','Longitude']]
        points=(location1.to_numpy(dtype ='float32'))
        xlist = [x[0] for x in points.tolist()]
        ylist = [x[1] for x in points.tolist()]
        xtranslated = [(x-min(xlist))*111*1000 for x in xlist]
        ytranslated = [(x-min(ylist))*111*1000 for x in ylist]
        pointzip = zip(xtranslated,ytranslated)
        points = np.array(list(pointzip))
        print("#### Find the Grid points under Micro Cell")
        stat_list_micro_inner = dict()
        for K in range(10,50):
#             micro_radius = 1000
            M =500
            micro_sites,f = mclp(points,K,micro_radius,M)
            microcoveredlist, not_coveredlist_micro, micro_sitedict = extract_coverage(micro_sites,points,micro_radius)
            ###################### rescaling ###########################
            microcoveredlist_rescaled = rescaled_points(xlist,ylist,np.array(microcoveredlist))
            not_coveredlist_micro_rescaled = rescaled_points(xlist,ylist,np.array(not_coveredlist_micro))
            micro_sitedict_rescaled = rescaled_points(xlist,ylist,np.array(list(micro_sitedict.values())))
            ############################################################
            stat_list_micro_inner[str(K)] = [len(microcoveredlist),len([k for k, v in dict(Counter([str(x) for x in microcoveredlist])).items() if v > 1]),microcoveredlist, not_coveredlist_micro, micro_sitedict,microcoveredlist_rescaled, not_coveredlist_micro_rescaled, micro_sitedict_rescaled]
        ##################################################################################################################
        df_micro = pd.DataFrame(list(stat_list_micro_inner.items()),columns = ['K_value','Values'])
        df_micro['Covered_area'] = [x[0] for x in df_micro.Values]
        df_micro['Redundant_area'] = [x[1] for x in df_micro.Values]
        df_micro = df_micro.iloc[:,np.r_[0,2:4]]
        df_micro['Cell'] = 'micro'
        df_micro['Area'] =  str(ctr)
        df_micro['No_of_points_to_optimize'] = len(points)
        df_micro['Weighted_value'] = (df_micro.Covered_area*2) - (df_micro.Redundant_area*5)
        K_micro = int(df_micro[df_micro['Weighted_value'] == df_micro.Weighted_value.quantile(q= .99, interpolation='lower')].K_value.min())
        df_micro['Best_K'] = 0
        df_micro.loc[(df_micro.K_value == str(K_micro)), "Best_K"] = 1
        stat_list_micro["area"+str(ctr)] = [stat_list_micro_inner, K_micro]
        ###################### getting micro cell centres for the best K ###########################
        microcoveredlist, not_coveredlist_micro, micro_sitedict = stat_list_micro_inner[str(K_micro)][2],stat_list_micro_inner[str(K_micro)][3],stat_list_micro_inner[str(K_micro)][4]
        print("K-value", K_micro)
        print("total points for optimization "+ str(len(points)))
        print("total points covered "+ str(len(microcoveredlist)))
        print("total points not covered "+ str(len(not_coveredlist_micro)))
        print("total no of circles needed to cover the points "+ str(len(micro_sitedict)))
        ##################################################################################################################
        if len(not_coveredlist_micro)>2 :
            print("\n")
            print("#### Find the Grid points under femto Cell")
            points = np.array(not_coveredlist_micro)
            print("points not covered by femto cells", len(points))
            stat_list_femto_inner = dict()
            for K in range(10,50):
#                 femto_radius = 300
                M =1000
                femto_sites,f = mclp(points,K,femto_radius,M)
                femtocoveredlist, not_coveredlist_femto, femto_sitedict = extract_coverage(femto_sites,points,femto_radius)
                ###################### rescaling ###########################
                femtocoveredlist_rescaled = rescaled_points(xlist,ylist,np.array(femtocoveredlist))
                not_coveredlist_femto_rescaled = rescaled_points(xlist,ylist,np.array(not_coveredlist_femto))
                femto_sitedict_rescaled = rescaled_points(xlist,ylist,np.array(list(femto_sitedict.values())))
                ############################################################
                stat_list_femto_inner[str(K)] = [len(femtocoveredlist),len([k for k, v in dict(Counter([str(x) for x in femtocoveredlist])).items() if v > 1]),femtocoveredlist, not_coveredlist_femto, femto_sitedict,femtocoveredlist_rescaled, not_coveredlist_femto_rescaled, femto_sitedict_rescaled]
            ##################################################################################################################
            df_femto = pd.DataFrame(list(stat_list_femto_inner.items()),columns = ['K_value','Values']) 
            df_femto['Covered_area'] = [x[0] for x in df_femto.Values]
            df_femto['Redundant_area'] = [x[1] for x in df_femto.Values]
            df_femto = df_femto.iloc[:,np.r_[0,2:4]]
            df_femto['Cell'] = 'femto'
            df_femto['Area'] =  str(ctr)
            df_femto['No_of_points_to_optimize'] = len(points)
            df_femto['Weighted_value'] = (df_femto.Covered_area*2) - (df_femto.Redundant_area*5)
            K_femto = int(df_femto[df_femto['Weighted_value'] == df_femto.Weighted_value.quantile(q= .99, interpolation='lower')].K_value.min())
            df_femto['Best_K'] = 0
            df_femto.loc[(df_femto.K_value == str(K_femto)), "Best_K"] = 1
            stat_list_femto["area"+str(ctr)] = [stat_list_femto_inner, K_femto]
            ###################### getting femto cell centres for the best K ###########################
            femtocoveredlist, not_coveredlist_femto, femto_sitedict = stat_list_femto_inner[str(K_femto)][2],stat_list_femto_inner[str(K_femto)][3],stat_list_femto_inner[str(K_femto)][4]
            print("K-value", K_femto)
            print("total points for optimization "+ str(len(points)))
            print("total points covered "+ str(len(femtocoveredlist)))
            print("total points not covered "+ str(len(not_coveredlist_femto)))
        else:
            femtocoveredlist = np.array([])
            not_coveredlist_femto = np.array([])
            femto_sites = np.array([])

            femtocoveredlist_rescaled = np.array([])
            not_coveredlist_femto_rescaled = np.array([])

            femto_radius = 0
            df_femto = pd.DataFrame(columns = df_micro.columns)
            K_femto = 0
            pass
        ##################################################################################################################
        df = df.append(pd.concat([df_micro, df_femto],                    
                              ignore_index = True,
                              sort = False))
    
    return df, stat_list_micro, stat_list_femto
```
